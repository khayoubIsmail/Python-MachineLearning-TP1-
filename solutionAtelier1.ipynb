```python
print("Enter deux entiers : ")
nbr1 = int(input("nombre1 = "))
nbr2 = int(input("nombre2 = "))
print("Avant le changement :\n nombre1 = {0:10} \n nombre2 = {1:10}".format(nbr1, nbr2))
tmp = nbr1
nbr1 = nbr2
nbr2 = tmp
print("Après le changement :\n nombre1 = {0:10} \n nombre2 = {1:10}".format(nbr1, nbr2))
```

# Exercice 3 :

```python
note1, coefficientNote1 = float(input("Donnez la premiere note ---> note1=")), int(input("Donnez le coefficient de premiere note ---> coefficientNote1="))
note2, coefficientNote2 = float(input("Donnez la deuxième note ---> note2=")), int(input("Donnez le coefficient de deuxième note ---> coefficientNote2="))
note3, coefficientNote3 = float(input("Donnez la toisème note ---> note3=")), int(input("Donnez le coefficient de toisème note ---> coefficientNote3="))
sommeCoefficients = coefficientNote1 + coefficientNote2 + coefficientNote3
moyenne = float(((note1*coefficientNote1)+(note2*coefficientNote2)+(note3*coefficientNote3))/sommeCoefficients)
print("\n la moyenne est égale : {0:.2f}".format(moyenne))
```
