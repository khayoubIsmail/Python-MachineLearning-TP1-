# Exercice 9 :

```python
print("En travail sur une fonction affine de format F(X)=AX+B \nDonnez les valeurs de A et B de votre choix:")
A = int(input("A="))
B = int(input("B="))
if B<0 :
    print("F(X)={0}X{1}".format(A, B))
elif B==0:
    print("F(X)={0}X".format(A))
else:
    print("F(X)={0}X+{1}".format(A, B))
n=int(input("Entrez la valeur de n : "))
a=int(input("Entrez la valeur de a : "))
b=int(input("Entrez la valeur de b : "))
h=(b-a)/n
Somme=0
# la méthode des rectangles :
for i in range(0,n):
    xi=a+(i*(h/2))
    xiplus1=a+((i+1)*(h/2))
    X=(xi+xiplus1)/2
    F_X=(A*X)+B
    Somme+=(xiplus1-xi)*F_X
print("la méthodes de réctangles --> I={0:.2f}".format(Somme))

# la méthode des trapèzes :
f_a=(A*a)+B
f_b=(A*b)+B
for i in range(1,n,2):
    xi=a+(i*(h/2))
    Somme+=(A*xi)+B
I=h*(((f_a+f_b)/2)+Somme)
print("la méthodes des trapèzes --> I={0:.2f}".format(I))
# la méthode de Simpson :
Somme1=0
for i in range(0,n,2):
    x_2i_plus_1=a+(((2*i)+1)*(h/2))
    F_X_2i_plus_1=(A*x_2i_plus_1)+B
    Somme1+=F_X_2i_plus_1
Somme2=0
for i in range(1,n,3):
    x_2i=a+((2*i)*(h/2))
    F_X_2_i = (A * x_2i) + B
    Somme2 += F_X_2_i
I=(h/6)*(f_a+f_b+(4*Somme1)+(2*Somme2))
print("la méthodes de Simpson --> I={0:.2f}".format(I))
```
