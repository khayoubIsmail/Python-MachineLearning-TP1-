### Dans R :

```python
from math import sqrt
a = float(input("Donnez la valeur de a : "))
if a==0:
    print("la valeur de a doit être non nulle")
else:
    b = float(input("Donnez la valeur de b : "))
    c = float(input("Donnez la valeur de c : "))
    Delta = (b*b)-(4*a*c)
    if Delta < 0:
        print("Aucun Sulutions dans R")
    elif Delta >0:
            x1 = (-b + sqrt(Delta)) / 2 * a
            x2 = (-b - sqrt(Delta)) / 2 * a
            print("l'équation a deux solutions :\n solution1={0:.2f}\n solution2={1:.2f}".format(x1, x2))
    else:
        x1 = -b/2*a
        print("l'équation a une seule solution :\n solution1={0:.2f}".format(x1))
```

```python
from math import sqrt
from cmath import *
a = float(input("Donnez la valeur de a : "))
if a==0:
    print("la valeur de a doit être non nulle")
else:
    b = float(input("Donnez la valeur de b : "))
    c = float(input("Donnez la valeur de c : "))
    Delta = (b**2)-(4*a*c)
    if Delta < 0:
        z1 = complex(-b/(2 * a), sqrt(-Delta)/(2 * a))
        z2 = complex(-b/(2 * a), -sqrt(-Delta)/(2 * a))
        print("l'équation a deux solutions Complexes :\n solution1={0:.2f}\n solution2={1:.2f}".format(z1, z2))
    elif Delta >0:
            x1 = (-b + sqrt(Delta)) / 2 * a
            x2 = (-b - sqrt(Delta)) / 2 * a
            print("l'équation a deux solutions :\n solution1={0:.2f}\n solution2={1:.2f}".format(x1, x2))
    else:
        x1 = -b/2*a
        print("l'équation a une seule solution :\n solution1={0:.2f}".format(x1))
```

```python
pSeuil,vSeuil = 2.3, 7.41
pressionCourant = float(input("Entrer la pression courant de l'enceinte : "))
volumeCourant = float(input("Entrer la pression courant de l'enceinte : "))
if pressionCourant > pSeuil:
    if volumeCourant > vSeuil:
        print("==> arrêt immédiat")
    else:
        print("==> augmenter le volume de l'enceinte")
elif volumeCourant > vSeuil:
        print("==> diminuer le volume de l'encient")
else:
    print("==>tout va bien")
```
